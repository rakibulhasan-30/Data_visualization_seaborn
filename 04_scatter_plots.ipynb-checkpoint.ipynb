# 04: Scatter Plots

This notebook demonstrates **scatter plots** and **regression plots** using Seaborn to analyze insurance data.  
We explore relationships between **BMI, Charges, and Smoking Status**.


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup Complete")
```

**Libraries Used:**
- `pandas` for data manipulation
- `matplotlib` and `%matplotlib inline` for plotting
- `seaborn` for scatter plots, regression lines, and advanced visualizations


```python
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
insurance_data.tail()
```

The insurance dataset contains features like **BMI**, **charges**, and **smoker status**.  
We will use scatter plots to explore the relationships between these variables.


```python
plt.figure(figsize=(12,8))
plt.title("Scatter Plot of BMI vs Charges")
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

A basic scatter plot to observe the relationship between **BMI** and **insurance charges**.


```python
plt.figure(figsize=(12,8))
plt.title("Regression Line Fit for BMI vs Charges")
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Adding a regression line helps visualize the **trend** between BMI and charges.


```python
plt.figure(figsize=(12,8))
plt.title("Scatter Plot of BMI vs Charges by Smoker Status")
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Using `hue='smoker'` to differentiate data points by **smoking status**, showing its impact on charges.


```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

A regression plot with separate lines for **smokers** and **non-smokers**.  
This highlights the effect of smoking on charges alongside BMI.


```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

A **swarm plot** visualizes individual data points for smokers vs non-smokers, showing distribution and spread of charges.


## Observations / Results

- BMI and charges are positively correlated; higher BMI often leads to higher insurance charges.  
- Smoking significantly increases charges, as shown by the separation of smoker vs non-smoker points.  
- Regression lines confirm that smokers generally have higher charges across all BMI values.  
- Swarm plot highlights individual variations and outliers in the dataset.
