# 03: Bar Charts and Heatmaps

This notebook demonstrates **bar charts** and **heatmaps** using Seaborn to visualize flight delays.  
We will analyze average arrival delays for different airlines by month.


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup Complete")
```

**Libraries Used:**
- `pandas` for data manipulation
- `matplotlib` for plotting
- `seaborn` for advanced visualizations like bar charts and heatmaps


```python
flight_data = pd.read_csv("flight_delays.csv", index_col="Month")
flight_data.head()
flight_data.tail()
```

The flight delays dataset is loaded with the **Month** column as the index.  
It contains average arrival delays (in minutes) for different airlines.


```python
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

A **bar chart** is used to visualize the average arrival delay of **Spirit Airlines (NK)** flights for each month.  
This helps to quickly identify months with higher delays.


```python
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for US Airlines, By Month")
sns.barplot(x=flight_data.index, y=flight_data['US'])
plt.ylabel("Arrival Delay (in minutes)")
```

Similarly, the average arrival delay for **US Airlines (US)** is plotted to compare with Spirit Airlines.


```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, By Month")
sns.heatmap(data=flight_data, annot=True)
plt.xlabel("Airlines")
```

A **heatmap** visualizes delays for all airlines by month in a single plot.  
The annotations (`annot=True`) show the exact delay values, making it easier to compare across airlines.


## Observations / Results

- Spirit Airlines (NK) has the highest delays in certain months, indicating seasonal peaks.  
- US Airlines (US) generally maintains lower delays, but some months show spikes.  
- The heatmap provides a quick overview of all airlines, highlighting which months and airlines experience higher delays.  
- Such visualizations help identify problem periods for airline scheduling and performance analysis.
