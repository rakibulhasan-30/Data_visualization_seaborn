# 07: Final Project - DC Comics Data Visualization

This notebook demonstrates a **comprehensive data visualization project** using the DC Comics dataset.  
We will explore character attributes like eye color, alignment, appearances, and powers through various visualizations.


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup Complete")
```

**Libraries Used:**
- `pandas` for data manipulation
- `matplotlib` for plotting
- `seaborn` for advanced visualizations


```python
dc_data = pd.read_csv("dc-wikia-data.csv", index_col="page_id")
dc_data.head()
```

The **DC Comics dataset** contains character information such as name, alignment, eye color, appearances, and powers.  
We set `page_id` as the index for easy reference.


```python
plt.figure(figsize=(12,8))
plt.title("Histogram: Eye Color Distribution by Name")
sns.histplot(x=dc_data['name'], y=dc_data['EYE'])
```

This histogram shows the distribution of **eye colors** among characters.  
It helps identify which eye colors are most common in DC characters.


```python
plt.figure(figsize=(10,6))
plt.title("Character Alignment Distribution")
sns.countplot(data=dc_data, x='ALIGN')
```

A **countplot** visualizes the number of characters by alignment (good, bad, neutral).  
This highlights the distribution of heroes and villains in the dataset.


```python
top_appearances = dc_data.sort_values(by='APPEARANCES', ascending=False).head(10)
plt.figure(figsize=(12,8))
sns.barplot(x='APPEARANCES', y='name', data=top_appearances)
plt.title("Top 10 Characters by Appearances")
```

This bar chart shows the **top 10 characters** with the most comic appearances.  
It helps identify the most prominent characters in DC Comics.


```python
plt.figure(figsize=(12,8))
plt.title("Scatter Plot: Eye Color vs Appearances")
sns.scatterplot(x='EYE', y='APPEARANCES', data=dc_data, hue='ALIGN')
```

A **scatter plot** showing the relationship between **eye color** and **appearances**, colored by alignment.  
This highlights any patterns between character features and their prominence.


```python
numeric_cols = ['APPEARANCES', 'YEAR']
plt.figure(figsize=(10,8))
sns.heatmap(dc_data[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

A **heatmap** shows correlation between numeric features like **appearances** and **year introduced**.  
It helps understand which attributes may be related.


## Observations / Results

- Eye color distribution is dominated by a few common colors among characters.  
- Most characters are aligned as **good**, fewer as **bad**, and very few are neutral.  
- Top 10 characters by appearances include iconic heroes and villains like Batman and Superman.  
- Scatterplots reveal some patterns between eye color, appearances, and alignment.  
- Correlation heatmap indicates weak relationships between numeric features, suggesting diverse character creation over time.
