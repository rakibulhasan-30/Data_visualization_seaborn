# 01: Hello Seaborn

This notebook demonstrates the basics of **Seaborn** for data visualization.  
We will load a dataset and create a simple line plot to visualize trends over time.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup Complete")
```

**Libraries Used:**
- `pandas` for data manipulation
- `matplotlib` and `%matplotlib inline` for plotting
- `seaborn` for advanced visualizations


```python
# Load FIFA dataset
fifa_data = pd.read_csv("fifa.csv", index_col="Date", parse_dates=True)
fifa_data.head()
```

We load the FIFA dataset using `pandas.read_csv()` and set the **Date** column as the index.  
`parse_dates=True` ensures that the dates are recognized as datetime objects.


```python
# Plot the data
plt.figure(figsize=(16,10))
sns.lineplot(data=fifa_data)
```

Here we use **Seaborn's `lineplot`** to visualize the trends of different columns over time.  
The figure size is set to 16x10 inches for better readability.
