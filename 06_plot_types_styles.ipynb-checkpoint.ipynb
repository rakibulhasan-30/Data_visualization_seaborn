# 06: Choosing Plot Types and Custom Styles

This notebook demonstrates how to **customize plot styles** and choose appropriate plot types using Seaborn.  
We will use Spotify streaming data to explore trends with different styles and enhancements.


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup Complete")
```

**Libraries Used:**
- `pandas` for data manipulation
- `matplotlib` for plotting
- `seaborn` for advanced visualizations and style customization


```python
spotify_data = pd.read_csv("spotify.csv", index_col="Date", parse_dates=True)
spotify_data.head()
```

We load the **Spotify dataset** with daily streams of popular songs, indexed by date.


```python
sns.set_style("darkgrid")  # Set the plot style
plt.figure(figsize=(16,10))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify_data)
```

We set a **darkgrid style** for better readability of line plots.  
The figure size is increased for clarity, and all songs are plotted together to observe overall trends.


```python
# Plot top 2 songs separately with different line styles
plt.figure(figsize=(16,10))
plt.title("Comparison of Top 2 Songs Streams in 2017-2018")
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You", linestyle='-', linewidth=2)
sns.lineplot(data=spotify_data['Despacito'], label="Despacito", linestyle='--', linewidth=2)
plt.xlabel("Date")
plt.ylabel("Daily Streams")

# Add another style
sns.set_style("whitegrid")
plt.figure(figsize=(16,10))
plt.title("Daily Streams with Whitegrid Style")
sns.lineplot(data=spotify_data[['Shape of You','Despacito']])
plt.xlabel("Date")
plt.ylabel("Daily Streams")
```

- Top 2 songs are plotted with **different line styles** to distinguish between them.  
- Line width and style are customized for clarity.  
- The second example uses a **whitegrid style** to show how changing the style affects the appearance of the plot.


## Observations / Results

- Customizing plot styles like `darkgrid` and `whitegrid` improves readability.  
- Using different line styles and widths helps distinguish multiple series in one plot.  
- Style choice depends on the type of data and the clarity needed for presentation.  
- Comparing top songs individually provides better insight than plotting all together.
